# **Resumen del caso**

El texto usa la historia de una cazatalentos que busca a “la mejor encestadora” de básquet para explicar el concepto de overfitting (sobreajuste) y fenómenos relacionados con la suerte, la maldición del ganador y el problema de las múltiples comparaciones.

- ***La Suerte:***

Variabilidad aleatoria inherente a todo proceso probabilístico.
Nadie encesta exactamente el mismo número de tiros en cada ronda, aunque tenga la misma habilidad promedio. Si se comparan muchas jugadoras parecidas, alguna puede ganar por suerte, aunque no sea realmente mejor. Es decir una jugadora si solo se la evalúa una vez, puede parecer mejor o peor de lo que realmente es.

En modelos predictivos ocurre lo mismo: un modelo puede tener un rendimiento muy alto por casualidad en un conjunto de datos, sin que eso signifique que sea realmente mejor.

- ***El problema de las múltiples comparaciones:***

Si hay 100 jugadoras con habilidades parecidas (índice de enceste entre 0.50 y 0.60),
es muy probable que alguna, solo por suerte, enceste un número inusualmente alto. Esa “ganadora” parece ser la mejor, pero en realidad fue la más afortunada.

En estadística:
si probás muchas hipótesis, alguna parecerá significativa aunque sea puro azar.

- ***La Maldicion del Ganador***

La ganadora del primer torneo (la que tuvo la mejor performance inicial)
no logra mantener su rendimiento en la siguiente prueba. Su alto desempeño inicial no reflejaba su verdadera habilidad, sino una combinación de talento + suerte.

El modelo que tuvo el mejor puntaje en el Public Leaderboard (la primera ronda)
suele caer en el Private Leaderboard (la segunda ronda).



**Que relacion tiene todo esto con el OVERFITTING ?**

Ocurre cuando un modelo (o una jugadora) parece rendir excelente durante la selección,
pero en realidad solo se adaptó a los detalles específicos de esa muestra o torneo, y no generaliza bien. Cuando la cazatalentos hace tirar 10 tiros a cada una de 100 jugadoras,
la ganadora suele ser una del “pelotón” (las de 0.50–0.60), no la verdadera mejor (0.70),
simplemente porque tuvo suerte y metió más tiros por azar.
 Cuando se la prueba de nuevo, su performance cae: eso es overfitting puro, entrenar un modelo que “memoriza” los datos de entrenamiento
y falla cuando se enfrenta a nuevos datos.

Pero entonces en el caso de las jugadoras de Basket, como podemos reducir el overfitting? Reduciendo la influencia de la suerte:
- Aumentando la cantidad de tiros por jugadora
- Teniendo una jugadora considerablemente mejor que las demas
- Comparar menos cantidad de jugadoras, ya que hay menos cantidad de jugadoras que pueden salir beneficiadas
- Volviendo a testear a las jugadoras, donde si una gano por suerte su rendimiento baja considerablemente

# Codigo:

Queremos ver quién es mejor, tiene mayor probabilidad de hacer más encestes.
Basandonos en que "Ci < Cj si la probabilidad de que Ci enceste menos que Cj es mayor que 0.5”.

A tener en cuenta:
- Estuve probando diferentes formas de reflejar la suerte en el codigo, por lo que se establecen dos codigos diferentes. El primero, en el que el incide de enceste que da la consigna se toma como en indice de enceste real que tiene la jugadora y el segundo con una penalidad por suerte dada por la cantidad de jugadoras.



```r

# SIN PENALIDAD POR SUERTE
# Datos reales del caso

cazatalentos <- data.frame(
  nombre = c("C1", "C2", "C3", "C4", "C5", "C6", "C7"),
  competidoras = c(100, 200, 100, 2, 1, 100, 101),
  tiros = c(100, 100, 100, 100, 100, 100, 100),
  indice_enceste = c(0.80, 0.80, 0.80, 0.80, 0.701, 0.79, 0.79)
)

cat("\n=== Datos base ===\n")
print(cazatalentos)

# ------------
# Simulaciones

set.seed(123)
nsim <- 10000   # número de simulaciones
n_tiros <- 100  # cada candidata hace 100 tiros

# Matriz para guardar los resultados de simulación
simulaciones <- matrix(NA, nrow = nsim, ncol = nrow(cazatalentos))
colnames(simulaciones) <- cazatalentos$nombre

# Uso distribución binomial: número de aciertos en 100 tiros con p = índice_enceste
for (i in 1:nrow(cazatalentos)) {
  simulaciones[, i] <- rbinom(nsim, n_tiros, cazatalentos$indice_enceste[i])
}

# -----------------------
# Comparaciones par a par

n_cand <- nrow(cazatalentos)
prob_matriz <- matrix(NA, nrow = n_cand, ncol = n_cand)
colnames(prob_matriz) <- rownames(prob_matriz) <- cazatalentos$nombre

# Calcular la probabilidad de que Ci enceste menos que Cj
for (i in 1:n_cand) {
  for (j in 1:n_cand) {
    prob_matriz[i, j] <- mean(simulaciones[, i] < simulaciones[, j])
  }
}

cat("\n=== Probabilidad de que Ci enceste menos que Cj (P(Ci < Cj)) ===\n")
print(round(prob_matriz, 3))

# --------------------------
# Ordenamiento final

# Contamos cuántas "derrotas" (prob > 0.5) tiene cada cazatalentos
cazatalentos$perdidas <- apply(prob_matriz > 0.5, 1, sum)

# Ordenamos según cantidad de derrotas (menos derrotas = mejor)
orden_final <- cazatalentos[order(cazatalentos$perdidas),
                            c("nombre", "indice_enceste", "perdidas")]

cat("\n=== ORDEN FINAL (de mejor a peor) ===\n")
print(orden_final)
```

Orden sin penalidad por suerte:
      
      C5 < C7 < C6 < C4 < C3 < C2 < C1

Curioso ver que si se toman esos encestes como los indices reales de enceste de cada jugadora el orden queda casi completamente al revez.

```r
# CON PENALIDAD POR SUERTE
# Datos reales del caso

cazatalentos <- data.frame(
  nombre = c("C1", "C2", "C3", "C4", "C5", "C6", "C7"),
  competidoras = c(100, 200, 100, 2, 1, 100, 101),
  tiros = c(100, 100, 100, 100, 100, 100, 100),
  indice_enceste = c(0.80, 0.80, 0.80, 0.80, 0.701, 0.79, 0.79)
)

cat("\n=== Datos base ===\n")
print(cazatalentos)

# ----------------------------------
# Ajuste simple por "penalidad por suerte"

# Creo un factor de penalidad según la cantidad de competidoras
# Entre 0 y 1, más competidoras -> mayor penalidad (reducción del índice)
# Para simplificar: peso = min(1, competidoras / 200)
# Este factor reduce un poco el índice de enceste. Multiplica por 0.1 para que el efecto sea no tan grande (10% máximo)

cazatalentos$indice_ajustado <- cazatalentos$indice_enceste * (1 - pmin(1, cazatalentos$competidoras / 500) * 0.1)


# Indices ajustados
cat("\n=== Indices: original vs ajustado por competidoras ===\n")
print(cazatalentos[, c("nombre", "competidoras", "indice_enceste", "indice_ajustado")])

# ------------
# Simulaciones

set.seed(123)
nsim <- 10000   # número de simulaciones
n_tiros <- 100  # cada candidata hace 100 tiros

# Matriz para guardar los resultados de simulación
simulaciones <- matrix(NA, nrow = nsim, ncol = nrow(cazatalentos))
colnames(simulaciones) <- cazatalentos$nombre

# Uso distribución binomial: número de aciertos en 100 tiros con p = índice_ajustado
for (i in 1:nrow(cazatalentos)) {
  simulaciones[, i] <- rbinom(nsim, n_tiros, cazatalentos$indice_ajustado[i])
}

# -----------------------
# Comparaciones par a par

n_cand <- nrow(cazatalentos)
prob_matriz <- matrix(NA, nrow = n_cand, ncol = n_cand)
colnames(prob_matriz) <- rownames(prob_matriz) <- cazatalentos$nombre

# Calcular la probabilidad de que Ci enceste menos que Cj
for (i in 1:n_cand) {
  for (j in 1:n_cand) {
    prob_matriz[i, j] <- mean(simulaciones[, i] < simulaciones[, j])
  }
}

cat("\n=== Probabilidad de que Ci enceste menos que Cj (P(Ci < Cj)) ===\n")
print(round(prob_matriz, 3))

# --------------------------
# Ordenamiento final

# Contamos cuántas "derrotas" (prob > 0.5) tiene cada cazatalentos
cazatalentos$perdidas <- apply(prob_matriz > 0.5, 1, sum)

# Ordenamos según cantidad de derrotas (menos derrotas = mejor)
orden_final <- cazatalentos[order(cazatalentos$perdidas),
                            c("nombre", "indice_enceste", "indice_ajustado", "perdidas")]

cat("\n=== ORDEN FINAL (de mejor a peor) con ajuste por competidoras ===\n")
print(orden_final)
```

      C5 < C2 < C7 < C6 < C3 < C1 < C4
