# **Resumen del caso**

El texto usa la historia de una cazatalentos que busca a “la mejor encestadora” de básquet para explicar el concepto de overfitting (sobreajuste) y fenómenos relacionados con la suerte, la maldición del ganador y el problema de las múltiples comparaciones.

- ***La Suerte:***

Variabilidad aleatoria inherente a todo proceso probabilístico.
Nadie encesta exactamente el mismo número de tiros en cada ronda, aunque tenga la misma habilidad promedio. Si se comparan muchas jugadoras parecidas, alguna puede ganar por suerte, aunque no sea realmente mejor. Es decir una jugadora si solo se la evalúa una vez, puede parecer mejor o peor de lo que realmente es.

En modelos predictivos ocurre lo mismo: un modelo puede tener un rendimiento muy alto por casualidad en un conjunto de datos, sin que eso signifique que sea realmente mejor.

- ***El problema de las múltiples comparaciones:***

Si hay 100 jugadoras con habilidades parecidas (índice de enceste entre 0.50 y 0.60),
es muy probable que alguna, solo por suerte, enceste un número inusualmente alto. Esa “ganadora” parece ser la mejor, pero en realidad fue la más afortunada.

En estadística:
si probás muchas hipótesis, alguna parecerá significativa aunque sea puro azar.

- ***La Maldicion del Ganador***

La ganadora del primer torneo (la que tuvo la mejor performance inicial)
no logra mantener su rendimiento en la siguiente prueba. Su alto desempeño inicial no reflejaba su verdadera habilidad, sino una combinación de talento + suerte.

El modelo que tuvo el mejor puntaje en el Public Leaderboard (la primera ronda)
suele caer en el Private Leaderboard (la segunda ronda).



**Que relacion tiene todo esto con el OVERFITTING ?**

Ocurre cuando un modelo (o una jugadora) parece rendir excelente durante la selección,
pero en realidad solo se adaptó a los detalles específicos de esa muestra o torneo, y no generaliza bien. Cuando la cazatalentos hace tirar 10 tiros a cada una de 100 jugadoras,
la ganadora suele ser una del “pelotón” (las de 0.50–0.60), no la verdadera mejor (0.70),
simplemente porque tuvo suerte y metió más tiros por azar.
 Cuando se la prueba de nuevo, su performance cae: eso es overfitting puro, entrenar un modelo que “memoriza” los datos de entrenamiento
y falla cuando se enfrenta a nuevos datos.

Pero entonces en el caso de las jugadoras de Basket, como podemos reducir el overfitting? Reduciendo la influencia de la suerte:
- Aumentando la cantidad de tiros por jugadora
- Teniendo una jugadora considerablemente mejor que las demas
- Comparar menos cantidad de jugadoras, ya que hay menos cantidad de jugadoras que pueden salir beneficiadas
- Volviendo a testear a las jugadoras, donde si una gano por suerte su rendimiento baja considerablemente

```r
# Datos de cada cazatalentos
cazatalentos <- data.frame(
  nombre = c("C1", "C2", "C3", "C4", "C5", "C6", "C7"),
  aciertos = c(80, 80, 80, 80, 701, 79, 79),
  tiros = c(100, 100, 100, 100, 1000, 100, 100)
)

# Valor esperado de aciertos en 100 tiros
cazatalentos$valor_esperado_100 <- (cazatalentos$aciertos / cazatalentos$tiros) * 100

# Orden final de mayor a menor valor esperado
orden <- cazatalentos[order(-cazatalentos$valor_esperado_100), c("nombre", "valor_esperado_100")]
cat("\nOrden de cazatalentos según valor esperado en 100 tiros:\n")
print(orden)


# --------------------------
# Comparaciones par a par

n_cand <- nrow(cazatalentos)
prob_matriz <- matrix(NA, n_cand, n_cand)
colnames(prob_matriz) <- rownames(prob_matriz) <- cazatalentos$nombre

for (i in 1:n_cand) {
  for (j in 1:n_cand) {
    # Probabilidad aproximada que Ci tenga menos encestes que Cj
    if (i == j) {
      prob_matriz[i,j] <- NA
    } else if (cazatalentos$valor_esperado_100[i] < cazatalentos$valor_esperado_100[j]) {
      prob_matriz[i,j] <- 0.75  # arbitrario > 0.5 si Ci < Cj
    } else {
      prob_matriz[i,j] <- 0.25  # arbitrario < 0.5 si Ci > Cj
    }
  }
}

cat("\nProbabilidad aproximada de que Ci tenga menos encestes que Cj:\n")
print(round(prob_matriz, 2))

```

```r
# ===============================
# Comparacion de Cazatalentos
# ===============================

set.seed(123)

# Datos de la consigna
# C1-C4: 80/100
# C5: 701/1000
# C6: 79/100
# C7: 79/100

cazatalentos <- data.frame(
  nombre = c("C1", "C2", "C3", "C4", "C5", "C6", "C7"),
  aciertos = c(80, 80, 80, 80, 701, 79, 79),
  tiros = c(100, 100, 100, 100, 1000, 100, 100)
)

# Calculo de la probabilidad posterior Beta(1,1)
cazatalentos$alfa <- 1 + cazatalentos$aciertos
cazatalentos$beta <- 1 + (cazatalentos$tiros - cazatalentos$aciertos)

# Valor esperado de aciertos en 100 tiros
cazatalentos$valor_esperado_100 <- 100 * (cazatalentos$alfa / (cazatalentos$alfa + cazatalentos$beta))

# Mostrar resultados
print(cazatalentos[, c("nombre", "valor_esperado_100")])

# --------------------------
# Comparaciones par a par
# --------------------------

# Simulacion simple para comparar prob(Ci < Cj)
nsim <- 10000
n_cand <- nrow(cazatalentos)
prob_matriz <- matrix(NA, n_cand, n_cand)
colnames(prob_matriz) <- rownames(prob_matriz) <- cazatalentos$nombre

for (i in 1:n_cand) {
  p_i <- rbeta(nsim, cazatalentos$alfa[i], cazatalentos$beta[i])
  tiros_i <- rbinom(nsim, 100, p_i)
  for (j in 1:n_cand) {
    p_j <- rbeta(nsim, cazatalentos$alfa[j], cazatalentos$beta[j])
    tiros_j <- rbinom(nsim, 100, p_j)
    prob_matriz[i, j] <- mean(tiros_i < tiros_j)
  }
}

cat("\nProbabilidad de que Ci tenga menos encestes que Cj (P(Ci < Cj)):\n")
print(round(prob_matriz, 3))

# --------------------------
# Orden final por valor esperado
# --------------------------
orden <- cazatalentos[order(-cazatalentos$valor_esperado_100), c("nombre", "valor_esperado_100")]
cat("\nOrden final (de mayor a menor valor esperado en 100 tiros):\n")
print(orden)
```

```r
# =========================================================================
# SCRIPT R CORREGIDO: JUSTIFICACIÓN DEL ORDENAMIENTO POR VALOR ESPERADO
#                     (Incorpora el factor N° de Comparaciones)
# =========================================================================

# 1. Definición de la Habilidad Real de las Jugadoras (Asunción de la Entrenadora)
#    Asumimos que todas las jugadoras que compitieron y obtuvieron 80/100
#    tenían una probabilidad real subyacente (p_real) de 0.70 (como en el enunciado
#    para ilustrar el overfitting). Este valor es arbitrario, pero esencial
#    para demostrar el efecto N.

TIROS <- 100
P_REAL_ASUMIDA <- 0.70 # Probabilidad subyacente asumida (70/100)
N_SIMULACIONES <- 5000

# 2. TABLA DE CANDIDATAS Y PARÁMETROS CLAVE --------------------------------

candidatas <- data.frame(
  nombre = c("C2", "C1", "C4", "C6", "C7", "C3", "C5"), # Orden predefinido para análisis
  N_competidoras = c(200, 100, 2, 1, 1, 1, 1), # El factor clave: N° de comparaciones (N)
  puntaje_observado = c(80, 80, 80, 79, 79, 80, 70.3), # C5 se usa con su real 70.3/100
  stringsAsFactors = FALSE
)

# 3. FUNCIÓN DE SIMULACIÓN PARA EL OVERFITTING (Maldición del Ganador)
#    Esta función mide el puntaje inflado por la suerte (overfitting) para cada N.

obtener_puntaje_ganador_promedio <- function(N, p_real, tiros, sims) {
  # Simulamos N_SIMULACIONES veces el torneo de N jugadoras.
  max_scores <- replicate(sims, {
    # Simula N jugadoras tirando 'tiros' veces con probabilidad 'p_real'.
    resultados <- rbinom(N, size = tiros, prob = p_real)
    return(max(resultados)) # Devolvemos el puntaje del ganador.
  })
  # El promedio de los puntajes ganadores (es lo que C1/C2/C4 observan)
  return(mean(max_scores))
}

# 4. CÁLCULO DE LA PENALIZACIÓN POR SUERTE (Factor de Overfitting)
#    La penalización es la diferencia entre el puntaje inflado y la p_real.

candidatas$puntaje_ganador_simulado <- NA
candidatas$penalizacion_suerte <- 0
candidatas$valor_esperado_estimado <- 0

for (i in 1:nrow(candidatas)) {
  N <- candidatas$N_competidoras[i]

  if (N > 1) {
    # a) Simular el puntaje promedio del ganador para este N (Overfitting)
    puntaje_max <- obtener_puntaje_ganador_promedio(N, P_REAL_ASUMIDA, TIROS, N_SIMULACIONES)

    # b) Calcular la penalización (el inflado que la suerte causa)
    penalizacion <- puntaje_max - (P_REAL_ASUMIDA * TIROS)

    # c) Aplicar la penalización al puntaje observado de 80
    VE_estimado <- candidatas$puntaje_observado[i] - penalizacion

    candidatas$puntaje_ganador_simulado[i] <- puntaje_max
    candidatas$penalizacion_suerte[i] <- round(penalizacion, 2)
    candidatas$valor_esperado_estimado[i] <- round(VE_estimado, 2)

  } else {
    # Para N=1 (C3, C6, C7, C5), no hay overfitting por comparación.
    # El valor esperado es simplemente el puntaje observado (o el promedio real conocido en C5).
    candidatas$valor_esperado_estimado[i] <- candidatas$puntaje_observado[i]
  }
}

# Reordenar por el Valor Esperado Estimado (p̂)
orden_final <- candidatas[order(candidatas$valor_esperado_estimado),
                          c("nombre", "N_competidoras", "puntaje_observado", "penalizacion_suerte", "valor_esperado_estimado")]

# 5. IMPRESIÓN DE LA JUSTIFICACIÓN FINAL ----------------------------------
cat("=================================================================\n")
cat("      JUSTIFICACIÓN DEL ORDEN POR RIESGO DE OVERFITTING (R)\n")
cat("=================================================================\n")
cat("Premisa: El puntaje real (Valor Esperado) es = Puntaje Observado - Penalización por Suerte.\n\n")

print(orden_final)

cat("\n-----------------------------------------------------------------\n")
cat("CONCLUSIÓN (Orden del Valor Esperado Estimado, del peor al mejor):\n")
cat("C2 (Máxima Penalización) < C1 < C5 (Habilidad real 70.3) < C4 < C6/C7 < C3 (Pura)\n")
cat("=================================================================\n")
```
